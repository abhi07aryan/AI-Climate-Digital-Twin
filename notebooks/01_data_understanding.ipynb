```python
import geopandas as gpd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10,10)
india = gpd.read_file("../data/raw/shapefiles/India_State_Boundary.shp")
india.head()
india.columns
india.plot(edgecolor = "black")
plt.title("Indian States")
plt.show()
up = india[india["State_Name"] == "Uttar Pradesh"]
up.plot(
    color="lightgreen",
    edgecolor="black"
)

plt.title("Uttar Pradesh")

plt.show()
up.to_file(
    "../data/raw/shapefiles/uttar_pradesh.shp"
)
up.geometry


```
